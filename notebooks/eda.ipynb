# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
# Load the data
data = pd.read_csv("../data/parkinsons/parkinsons.data")
data.head()
```

```python
print(f"Shape of the data : {data.shape}")
```

```python
data.info()
```

```python
data.describe()
```

```python
if data.isna().sum().sum() == 0:
    print("There are no missing values")
    
        
if data.duplicated().sum() == 0:
    print("There are no duplicated values")    
```

```python
# Analyse the target, it is imbalanced
data['status'].value_counts()
```

```python
# Visualisation
data.hist(figsize=(16, 12))
plt.show()
```

```python
# data without name
data_wn = data.drop(columns="name")
# data without name and status
data_wns = data.drop(columns=["name", "status"])
```

```python
# Corroletion heatmap
sns.heatmap(data_wn.corr(), annot=False, cmap="coolwarm", center=0)
plt.title("Correlation matrix")
plt.show()
```

```python
# Search for outliers
for col in ["MDVP:Jitter(%)", "MDVP:Shimmer", "spread1", "DFA"]:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
Q1 = data_wns.quantile(0.25)
Q3 = data_wns.quantile(0.75)

IQR = Q3 - Q1
```

```python
outliers = ((data_wns < (Q1 - 1.5 * IQR)) | (data_wns > (Q3 + 1.5 * IQR)))

outlier_counts = outliers.sum().sort_values(ascending=False)
outlier_counts.head(10)
```

```python
# PCA
pca = PCA(n_components=3)
pca.fit(data_wns)
```

```python
# 3 components gives us more than 99.8% of information
print(pca.explained_variance_ratio_.sum())
pca.explained_variance_ratio_
```
